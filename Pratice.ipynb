```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# 1. line plot.

```python
x = range(1,50)
y = [value * 3 for value in x]
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
plt.show()
```

# 2.Normal plot.

```python
x = [1,2,3]
y = [2,4,1]
plt.plot(x,y)
plt.xlabel("x - label.")
plt.ylabel('y - label.')
plt.show()
```

```python
x = [10,20,30]
y = [20,40,10]
plt.plot(x, y, label = "line - 1")
x = [10,20,30]
y = [40,10,30]
plt.plot(x, y, label = 'line - 2')
plt.xlabel('x - axis')
plt.ylabel('y - label')
plt.legend()
plt.title('Two or more lines on same plot with suitable legends')
plt.show()
```

# 5. Graph for gave width and colour.

```python
x = [10,20,30]
y = [20,40,10]
plt.plot(x, y, label = "line - 1 width - 3", color = "red", linewidth = 3)
x = [10,20,30]
y = [40,10,30]
plt.plot(x, y, label = 'line - 2 width - 5',color = "blue", linewidth = 5)
plt.xlabel('x - axis')
plt.ylabel('y - label')
plt.legend()
plt.title('Two or more lines on same plot with suitable legends')
plt.show()
```
